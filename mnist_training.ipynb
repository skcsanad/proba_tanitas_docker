```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
# Loading data
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
# Onehot encoding
y_train_1h = tf.keras.utils.to_categorical(y_train)
y_test_1h = tf.keras.utils.to_categorical(y_test)
```

```python
# Building model
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Conv2D(8, (4,4), activation='relu', input_shape=(28, 28, 1)))
model.add(tf.keras.layers.Dropout(0.25))
model.add(tf.keras.layers.Dense(64, activation='relu'))
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(10, activation='softmax'))
model.compile('Adam', loss='CategoricalCrossentropy', metrics=['accuracy', 'categorical_crossentropy'])
model.summary()
```

```python
# Defining callbacks
es = tf.keras.callbacks.EarlyStopping(patience=2, monitor='val_loss', verbose=1)
mc = tf.keras.callbacks.ModelCheckpoint(filepath='best_model.h5', save_best_only=True, verbose=1)
```

```python
history = model.fit(x_train, y_train_1h, epochs=10, validation_data=(x_test, y_test_1h), callbacks=[es, mc])
```

```python
#Plotting training history
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()


plt.plot(history.history['categorical_crossentropy'])
plt.plot(history.history['val_categorical_crossentropy'])
plt.title('Categorical Crossentropy Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
```
